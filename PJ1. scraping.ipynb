```python
from selenium import webdriver
from selenium.webdriver import ActionChains
```

```python
# data 쌓는 함수 생성
def get_item_list():
    result_list= []
    results = ch_driver.find_elements_by_xpath(item_xpath)
    for item in results:
        idx = item.find_element_by_class_name('bl_no').text
        subject = item.find_element_by_class_name('bl_subject').text
        category = item.find_element_by_class_name('bl_category').text
        name = item.find_element_by_class_name('bl_name').text
        date = item.find_element_by_class_name('bl_date').text
        agree = item.find_element_by_class_name('bl_agree').text
        print(idx, subject, category, name, date, agree, sep='\n')
        result_list.append((
        idx,
        subject,
        category,
        name,
        date,
        agree,
        ))
    return result_list
```

```python
# paging - ActionChains로 구현
CSS_btn_next = ".p_btn.next"
def go_next_page():
    next_btn = ch_driver.find_element_by_css_selector(CSS_btn_next)
    ActionChains(ch_driver).click(next_btn).perform()
```

```python
go_next_page() # 다음페이지 잘 넘어가는 부분 확인
```

```python
from time import sleep

ch_driver.implicitly_wait(5)

all_result = []
for i in range(1, 30+1):
    item_data = get_item_list()
    all_result.append(tuple(item_data))
    go_next_page()
    sleep(2)
```

## 느낀 점

### 어려웠던 점

1. 처음에 find_element...를 써서 큰 리스트를 잡으려 했지만, 다른 리스트임에도 같은 id없이 같은 class명을 쓰고, 또 내부 세부 리스트가 잡히지 않아 실패.. -> 결국 Xpath로 리스트 잡음
2. paging을 일으키는 버튼을 잡으려는데 javascript가 보이지 않아, 수업에서 배웠던 부분을 활용할 수 없었음 -> 구글링 후 ActionChains를 활용하여 잡음
